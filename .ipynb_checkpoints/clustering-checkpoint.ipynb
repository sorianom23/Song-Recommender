## Clustering

```python
import pandas as pd
from sklearn import datasets 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Load data
concated_nothot = pd.read_csv('/Users/mariasoriano/Desktop/Ironhack/lab-web-scraping-single-page/final_concated_nothot.csv')
concated_hot100 = pd.read_csv('/Users/mariasoriano/Desktop/Ironhack/lab-web-scraping-single-page/concated_hot100_final')
```

```python
# Concat both dfs
df = pd.concat([concated_hot100, concated_nothot], ignore_index=True)
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
```

```python
#let's drop the cols we don't need for the clustering because they are not relevant
#artist, title, id, type, uri, track_href, analysis_url
X = df.drop(columns=['title', 'artist', 'id', 'type', 'uri', 'track_href', 'analysis_url'], axis=1)
```

```python
X.replace("Null", 0, inplace=True)
```

```python
#making sure all types of X are numeric in order to do the StandardScaler()
X = X.apply(pd.to_numeric)
```

### Apply StandardScaler()

```python
#Save with a pickle

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
#display(X.head())
#print()
display(X_scaled_df.head())
```

```python
filename = "src/scalers/scaler.pkl"
with open(filename, "wb") as file: 
    pickle.dump(scaler,file)
```

## Clustering the songs with K-Means

```python
kmeans = KMeans(n_clusters=8, random_state=1234)
kmeans.fit(X_scaled_df)
```

```python
# Check how many observations for each cluster
clusters = kmeans.predict(X_scaled_df)
pd.Series(clusters).value_counts().sort_index()
```

```python
# for display purposes:
X["cluster"] = clusters
X.head()
```

```python
kmeans.inertia_
```

## Using the Elbow Method for chosing the value of K

```python
import pickle

K = range(2, 21)
inertia = [] 
silhouette = [] 

for k in K:
    print("Training a K-Means model with {} neighbours! ".format(k)) 
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                   verbose=1)
                    
    kmeans.fit(X_scaled_df)
    filename = "/Users/mariasoriano/Desktop/Ironhack/APIs" + str(k) + ".pickle" 
    with open(filename, "wb") as file: 
        pickle.dump(kmeans,file)
    inertia.append(kmeans.inertia_) 
    silhouette.append(silhouette_score(X_scaled_df, kmeans.predict(X_scaled_df)))


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```

**3, 9, 14 and 16** look like possible optimal **values for K**
